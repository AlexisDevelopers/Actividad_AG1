##Devolución de cambio por técnica voraz

```
def cambio_monedas(N, SM):
  SOLUCION = [0]*len(SM)   #SOLUCION = [0,0,0,0,..]
  ValorAcumulado = 0

  for i,valor in enumerate(SM):
    monedas =  (N-ValorAcumulado)//valor
    SOLUCION[i] = monedas
    ValorAcumulado = ValorAcumulado + monedas*valor

    if ValorAcumulado == N:
      return SOLUCION


cambio_monedas(15,[25,10,5,1])
```

## N-Reinas por técnica de vueta atrás


```
def escribe(S):
  n = len(S)
  for x in range(n):
    print("")
    for i in range(n):
      if S[i] == x+1:
        print(" X " , end="")
      else:
        print(" - ", end="")


def es_prometedora(SOLUCION,etapa):
  #print(SOLUCION)
  #Si la solución tiene dos valores iguales no es valida => Dos reinas en la misma fila
  for i in range(etapa+1):
    #print("El valor " + str(SOLUCION[i]) + " está " +  str(SOLUCION.count(SOLUCION[i])) + " veces")
    if SOLUCION.count(SOLUCION[i]) > 1:
      return False

    #Verifica las diagonales
    for j in range(i+1, etapa +1 ):
      #print("Comprobando diagonal de " + str(i) + " y " + str(j))
      if abs(i-j) == abs(SOLUCION[i]-SOLUCION[j]) : return False
  return True



def reinas(N, solucion=[], etapa=0):
  if len(solucion) == 0:
      solucion=[0 for i in range(N)]

  for i in range(1, N+1):
    solucion[etapa] = i

    if es_prometedora(solucion, etapa):
      if etapa == N-1:
        print(solucion)
        #escribe(solucion)
        print()
      else:
        reinas(N, solucion, etapa+1)
    else:
      None

    solucion[etapa] = 0

reinas(8)
```

## Viaje por el rio. Programación dinámica

```
TARIFAS = [
[0,5,4,3,999,999,999],
[999,0,999,2,3,999,11],
[999,999, 0,1,999,4,10],
[999,999,999, 0,5,6,9],
[999,999, 999,999,0,999,4],
[999,999, 999,999,999,0,3],
[999,999,999,999,999,999,0]
]



################################################################
def Precios(TARIFAS):
################################################################
  #Total de Nodos
  N = len(TARIFAS[0])

  #Inicialización de la tabla de precios
  PRECIOS = [ [9999]*N for i in [9999]*N]
  RUTA = [ [""]*N for i in [""]*N]

  for i in range(0,N-1):
    RUTA[i][i] = i             #Para ir de i a i se "pasa por i"
    PRECIOS[i][i] = 0          #Para ir de i a i se se paga 0
    for j in range(i+1, N):
      MIN = TARIFAS[i][j]
      RUTA[i][j] = i

      for k in range(i, j):
        if PRECIOS[i][k] + TARIFAS[k][j] < MIN:
            MIN = min(MIN, PRECIOS[i][k] + TARIFAS[k][j] )
            RUTA[i][j] = k          #Anota que para ir de i a j hay que pasar por k
        PRECIOS[i][j] = MIN

  return PRECIOS,RUTA
################################################################


PRECIOS,RUTA = Precios(TARIFAS)
#print(PRECIOS[0][6])

print("PRECIOS")
for i in range(len(TARIFAS)):
  print(PRECIOS[i])

print("\nRUTA")
for i in range(len(TARIFAS)):
  print(RUTA[i])

#Determinar la ruta con Recursividad
def calcular_ruta(RUTA, desde, hasta):
  if desde == hasta:
    #print("Ir a :" + str(desde))
    return ""
  else:
    return str(calcular_ruta( RUTA, desde, RUTA[desde][hasta])) +  \
                ',' + \
                str(RUTA[desde][hasta] \
              )

print("\nLa ruta es:")
calcular_ruta(RUTA, 0,6)
```

## 1D — Fuerza bruta

Probamos todas las parejas y nos quedamos con la distancia mínima.
Complejidad temporal: **O(n²)** (todas las parejas).
Complejidad espacial: **O(1)** adicional.

```
def closest_pair_1d_bruteforce(nums: List[int]) -> Tuple[Tuple[int, int], int]:
    """
    Devuelve: ((a, b), distancia)
    Distancia en 1D = |a - b|
    """
    n = len(nums)
    if n < 2:
        raise ValueError("Se requieren al menos 2 puntos.")

    best_pair = (nums[0], nums[1])
    best_dist = abs(nums[0] - nums[1])

    for i in range(n):
        for j in range(i + 1, n):
            d = abs(nums[i] - nums[j])
            if d < best_dist:
                best_dist = d
                best_pair = (nums[i], nums[j])

    return best_pair, best_dist


pair_1d_bf, dist_1d_bf = closest_pair_1d_bruteforce(LISTA_1D)
pair_1d_bf, dist_1d_bf
```

## 1D — Mejora: Divide y vencerás

Idea:
1. Ordenar la lista.
2. Aplicar Divide y Vencerás: resolver en mitades y combinar.
En 1D, una vez ordenado, el mínimo siempre está entre **elementos consecutivos**,
así que la combinación es muy simple.

Complejidad:
- Ordenar: **O(n log n)**
- Escaneo de vecinos: **O(n)**
Total: **O(n log n)** (mejora frente a O(n²)).
Espacio adicional: depende de implementación, típico **O(n)** por copias/recursión (o menor si se hace in-place).

```
def closest_pair_1d_divide_conquer(nums: List[int]) -> Tuple[Tuple[int, int], int]:
    """
    En 1D: ordenar y revisar pares consecutivos.
    Esto es equivalente a D&V porque el mínimo global queda preservado al combinar.
    """
    if len(nums) < 2:
        raise ValueError("Se requieren al menos 2 puntos.")

    arr = sorted(nums)  # O(n log n)

    best_pair = (arr[0], arr[1])
    best_dist = abs(arr[1] - arr[0])

    for i in range(1, len(arr) - 1):
        d = abs(arr[i + 1] - arr[i])
        if d < best_dist:
            best_dist = d
            best_pair = (arr[i], arr[i + 1])

    return best_pair, best_dist


pair_1d_dc, dist_1d_dc = closest_pair_1d_divide_conquer(LISTA_1D)
pair_1d_dc, dist_1d_dc
```

## 2D — Fuerza bruta

Distancia euclídea: $d((x_1,y_1),(x_2,y_2))=\sqrt{(x_1-x_2)^2+(y_1-y_2)^2}$

- Complejidad temporal: $O(n^2)$  
- Complejidad espacial: $O(1)$ adicional

```
Point2D = Tuple[int, int]

def dist_2d(a: Point2D, b: Point2D) -> float:
    return math.hypot(a[0] - b[0], a[1] - b[1])

def closest_pair_2d_bruteforce(points: List[Point2D]) -> Tuple[Tuple[Point2D, Point2D], float]:
    n = len(points)
    if n < 2:
        raise ValueError("Se requieren al menos 2 puntos.")

    best_pair = (points[0], points[1])
    best_dist = dist_2d(points[0], points[1])

    for i in range(n):
        for j in range(i + 1, n):
            d = dist_2d(points[i], points[j])
            if d < best_dist:
                best_dist = d
                best_pair = (points[i], points[j])

    return best_pair, best_dist


pair_2d_bf, dist_2d_bf = closest_pair_2d_bruteforce(LISTA_2D)
pair_2d_bf, dist_2d_bf
```

## 2D — Divide y vencerás (Closest Pair clásico)

Algoritmo clásico:
1. Ordenar por $x$ (y opcionalmente por $y$).
2. Dividir en dos mitades.
3. Resolver recursivamente cada mitad.
4. Combinar: revisar solo puntos cercanos a la línea de partición (*strip*) ordenados por $y$.

Complejidad típica: $O(n\log n)$  
(Se obtiene por recurrencia $T(n)=2T(n/2)+O(n)$ con combinación eficiente.)

```
def closest_pair_2d_divide_conquer(points: List[Point2D]) -> Tuple[Tuple[Point2D, Point2D], float]:
    if len(points) < 2:
        raise ValueError("Se requieren al menos 2 puntos.")

    px = sorted(points, key=lambda p: p[0])
    py = sorted(points, key=lambda p: p[1])

    def rec(px: List[Point2D], py: List[Point2D]) -> Tuple[Tuple[Point2D, Point2D], float]:
        n = len(px)
        if n <= 3:
            return closest_pair_2d_bruteforce(px)

        mid = n // 2
        left_x = px[:mid]
        right_x = px[mid:]
        mid_x = px[mid][0]

        left_set = set(left_x)
        left_y = [p for p in py if p in left_set]
        right_y = [p for p in py if p not in left_set]

        (pair_left, d_left) = rec(left_x, left_y)
        (pair_right, d_right) = rec(right_x, right_y)

        best_pair, best_d = (pair_left, d_left) if d_left <= d_right else (pair_right, d_right)

        # Strip: puntos con |x - mid_x| < best_d
        strip = [p for p in py if abs(p[0] - mid_x) < best_d]

        # En 2D basta revisar un número acotado de vecinos en strip (clásico: hasta 7)
        for i in range(len(strip)):
            for j in range(i + 1, min(i + 8, len(strip))):
                d = dist_2d(strip[i], strip[j])
                if d < best_d:
                    best_d = d
                    best_pair = (strip[i], strip[j])

        return best_pair, best_d

    return rec(px, py)


pair_2d_dc, dist_2d_dc = closest_pair_2d_divide_conquer(LISTA_2D)
pair_2d_dc, dist_2d_dc
```

## 3D — Extensión

Distancia euclídea 3D: $d((x_1,y_1,z_1),(x_2,y_2,z_2))=\sqrt{(x_1-x_2)^2+(y_1-y_2)^2+(z_1-z_2)^2}$

Primero implementamos fuerza bruta ($O(n^2)$).  
Nota: existen mejoras tipo Divide y Vencerás / KD-Tree para bajar el coste, pero para la extensión 3D una implementación robusta completa es más larga; la fuerza bruta valida el concepto.

```
Point3D = Tuple[int, int, int]

def dist_3d(a: Point3D, b: Point3D) -> float:
    return math.sqrt((a[0]-b[0])**2 + (a[1]-b[1])**2 + (a[2]-b[2])**2)

def closest_pair_3d_bruteforce(points: List[Point3D]) -> Tuple[Tuple[Point3D, Point3D], float]:
    n = len(points)
    if n < 2:
        raise ValueError("Se requieren al menos 2 puntos.")

    best_pair = (points[0], points[1])
    best_dist = dist_3d(points[0], points[1])

    for i in range(n):
        for j in range(i + 1, n):
            d = dist_3d(points[i], points[j])
            if d < best_dist:
                best_dist = d
                best_pair = (points[i], points[j])

    return best_pair, best_dist


pair_3d_bf, dist_3d_bf = closest_pair_3d_bruteforce(LISTA_3D)
pair_3d_bf, dist_3d_bf
```

## Resumen de complejidades

- 1D Fuerza bruta: **O(n²)**  
- 1D Mejorado (ordenar + vecinos): **O(n log n)**

- 2D Fuerza bruta: **O(n²)**  
- 2D Divide y vencerás clásico: **O(n log n)**

- 3D Fuerza bruta: **O(n²)**
